```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.interpolate as spline
```

```python
dataset1 = pd.read_csv('dataQ1.csv')
dataset1
```

```python
tamanhoDataset1 = list(dataset1.iloc[:,0].values)
sequencialDataset1 = list(dataset1.iloc[:,1].values)
distribuidaDataset1 = list(dataset1.iloc[:,2].values)
speedupDataset1 = list(dataset1.iloc[:,3].values)
```

```python
plt.figure()
plt.plot(tamanhoDataset1,sequencialDataset1,"ro-",label="Sequencial")
plt.plot(tamanhoDataset1,distribuidaDataset1,"go-",label="Distribuida")
plt.rcParams['figure.figsize'] = (20,10)
plt.title("Tamanho x (Sequencial x Distribuida)")
plt.ylabel("Tamanho")
plt.xlabel("Tempos")
plt.legend()
plt.grid(True)
plt.show()
```
